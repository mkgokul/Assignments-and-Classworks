# Who scores the most on average for math, reading and writing based on 
# Gender
# Test preparation course


```python
df
```

```python
df1=df.groupby(['gender']).mean()
df2=pd.DataFrame(df1).reset_index()
df2
```

```python
l=np.arange(2)
plt.figure(figsize=(12,8))

plt.bar(l-.17,df2['math score'],color='yellow',width=.17,edgecolor='black')
plt.bar(l,df2['reading score'],color='red',width=.17,edgecolor='black')
plt.bar(l+.17,df2['writing score'],color='blue',width=.17,edgecolor='black')
plt.xticks(l,df2['gender'])

plt.title("Avg scores, Gender wise for each sections",fontweight='bold',fontsize=14)
plt.xlabel('Gender',fontsize=14)
plt.ylabel('Avg score',fontsize=14)
plt.grid(linestyle='--',linewidth=.8)
plt.legend(loc="upper center",labels=('Math','Reading','Writing'))
```

```python
#Most scored gender in:-
#Math-male
#Reading-male
#writing-Female
```

```python
df3=df.groupby(['test preparation course']).mean()
df3=pd.DataFrame(df3).reset_index()
df3
```

```python
l=np.arange(2)
plt.figure(figsize=(12,8))

plt.bar(l-.17,df3['math score'],color='yellow',width=.17,edgecolor='black')
plt.bar(l,df3['reading score'],color='red',width=.17,edgecolor='black')
plt.bar(l+.17,df3['writing score'],color='blue',width=.17,edgecolor='black')
plt.xticks(l,df3['test preparation course'])

plt.title("Avg scores, Preparation course wise for each sections",fontweight='bold',fontsize=14)
plt.xlabel('Preparation course',fontsize=14)
plt.ylabel('Avg score',fontsize=14)
plt.grid(linestyle='--',linewidth=.8)
plt.legend(loc="upper center",labels=('Math','Reading','Writing'))
```

```python
#Most scored test preparation wise:-
#Math-completed
#Reading-completed+
#writing-writing
```

# What do you think about the scoring variation for math, reading and writing
# based on
â— Gender
â— Test preparation course

```python
df4=df.groupby(['gender']).std()
df4=pd.DataFrame(df4).reset_index()
df4
```

```python
l=np.arange(2)
plt.figure(figsize=(12,8))

plt.bar(l-.17,df4['math score'],color='yellow',width=.17,edgecolor='black')
plt.bar(l,df4['reading score'],color='red',width=.17,edgecolor='black')
plt.bar(l+.17,df4['writing score'],color='blue',width=.17,edgecolor='black')
plt.xticks(l,df4['gender'])

plt.title("std of scores, Gender wise for each sections",fontweight='bold',fontsize=14)
plt.xlabel('Gender',fontsize=14)
plt.ylabel('Std',fontsize=14)
plt.grid(linestyle='--',linewidth=.8)
plt.legend(loc="upper center",labels=('Math','Reading','Writing'))
```

```python
#Most scoring variation gender wise:-
#Math-Female
#Reading-Female+
#writing-FEmale

#Females have most deviation from the mean score
```

```python
df5=df.groupby(['test preparation course']).std()
df5=pd.DataFrame(df5).reset_index()
df5
```

```python
l=np.arange(2)
plt.figure(figsize=(12,8))

plt.bar(l-.17,df5['math score'],color='yellow',width=.17,edgecolor='black')
plt.bar(l,df5['reading score'],color='red',width=.17,edgecolor='black')
plt.bar(l+.17,df5['writing score'],color='blue',width=.17,edgecolor='black')
plt.xticks(l,df5['test preparation course'])

plt.title("Std of scores, Preparation course wise for each sections",fontweight='bold',fontsize=14)
plt.xlabel('test preparation course',fontsize=14)
plt.ylabel('Std',fontsize=14)
plt.grid(linestyle='--',linewidth=.8)
plt.legend(loc="upper center",labels=('Math','Reading','Writing'))
```

```python
#Most scoring variation test preparation wise:-
#Math-NONE
#Reading-NONE
#writing-NONE

#Those having test prepration course as NONE have most deviation from the mean score
```
