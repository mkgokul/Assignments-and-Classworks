#  Examine the data by using visualizations.

```python
data1 = data.groupby(["Region"]).sum()
data1
```

```python
visited=[]
for x in data['Region']:
    if x not in visited:
        visited.append(x)
len(visited)        
```

```python
l=np.arange(len(visited))
l
```

```python


plt.figure(figsize=(12,6))
plt.bar(l,data1['Population in thousands (2017)'],color='maroon',width=1,edgecolor='black')
plt.xticks(l,visited,rotation=90)
plt.xlabel('Region',fontsize=14)
plt.ylabel('Population in thousands',fontsize=14)
plt.title('Population in different Regions',fontweight='bold',fontsize=14)
plt.grid(linestyle='--',linewidth=.8)
```

```python
plt.figure(figsize=(12,6))
plt.bar(l,data1['GDP per capita (current US$)'],color='blue',width=1,edgecolor='black')
plt.xticks(l,visited,rotation=90)
plt.xlabel('Region',fontsize=14)
plt.ylabel('GDP per capita (current US$)',fontsize=14)
plt.title('GDP in different Regions',fontweight='bold',fontsize=14)
plt.grid(linestyle='--',linewidth=.8)
```

```python
plt.figure(figsize=(12,6))
plt.bar(l,data1['International trade: Imports (million US$)'],color='cyan',width=1,edgecolor='black')
plt.xticks(l,visited,rotation=90)
plt.xlabel('Region',fontsize=14)
plt.ylabel('International trade: Imports (million US$)',fontsize=14)
plt.title('International trade: Imports in different Regions',fontweight='bold',fontsize=14)
plt.grid(linestyle='--',linewidth=.8)
```

# On average which region has the highest
a. GDP per capita (current US$)
b. International trade: Imports (million US$)

```python
data['Surface area (km2)'].describe()['mean']
```

```python
data1 = data.groupby(["Region"]).mean()
data1
```

```python
print('Region with highest GDP is ')
t=data1['GDP per capita (current US$)'].max()
data1[data1['GDP per capita (current US$)']==t]
```

```python
print('Region with highest Trade is ')
q=data1['International trade: Imports (million US$)'].max()
data1[data1['International trade: Imports (million US$)']==q]
```

# Find out which region has more consistent with respect to GDP per capita (current US$)


```python
gdp=data.groupby(["Region"]).describe()
gdpData=gdp['GDP per capita (current US$)']
c=gdpData['std'].min()
print('Region having more consistency with respect to GDP')
gdpData[gdpData['std']==c]
```
